```python
import tensorflow as tf
import numpy
from openpyxl import load_workbook

xlsx = load_workbook(filename='data.xlsx')
data = xlsx['Sheet']

x_data, y_data = [], []
for i in data.rows:
    if i[0].value < i[1].value:
        x_data.append([i[0].value, i[1].value])
    else:
        x_data.append([i[1].value, i[0].value])
        
    y_data.append([i[2].value])
    
y_data_one_hot = tf.one_hot(y_data, 3)

x_data = numpy.array(x_data)
y_data = numpy.array(y_data)
```

```python
epochs = 50000

import tensorflow as tf
import numpy

X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

w1 = tf.Variable(tf.random_normal([2, 50]))
b1 = tf.Variable(tf.random_normal([50]))
layer_1 = tf.nn.relu(tf.matmul(X, w1) + b1)

w2 = tf.Variable(tf.random_normal([50, 100]))
b2 = tf.Variable(tf.random_normal([100]))
layer_2 = tf.nn.relu(tf.matmul(layer_1, w2) + b2)

w3 = tf.Variable(tf.random_normal([100, 3]))
b3 = tf.Variable(tf.random_normal([3]))
layer_3 = tf.matmul(layer_2, w3) + b3

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=layer_3, labels=y_data_one_hot))
optimizer = tf.train.GradientDescentOptimizer(0.038).minimize(cost) #In my project 0.038 is best value of this model.

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(epochs):
        if step % 100 == 0:
            print("Step : %d Cost : %f" % (step, sess.run([cost, optimizer], feed_dict={X:x_data, Y:y_data})[0]))
            
    print("Final Cost is %f" % sess.run(cost, feed_dict={X:x_data, Y:y_data}))
```

```python
def KBO_num(name):
    if(name == "넥센" or name == "키움"):
        return 0
    elif(name == "두산"):
        return 1
    elif(name == "롯데"):
        return 2
    elif(name == "삼성"):
        return 3
    elif(name == "한화"):
        return 4
    elif(name == "KIA"):
        return 5
    elif(name == "KT"):
        return 6
    elif(name == "NC"):
        return 7
    elif(name == "LG"):
        return 8
    elif(name == "SK"):
        return 9
    else:
        return 10
    
def num_to_team(num):
    if(num == 0):
        return "넥센"
    elif(num == 1):
        return "두산"
    elif(num == 2):
        return "롯데"
    elif(num == 3):
        return "삼성"
    elif(num == 4):
        return "한화"
    elif(num == 5):
        return "KIA"
    elif(num == 6):
        return "KT"
    elif(num == 7):
        return "NC"
    elif(num == 8):
        return "LG"
    elif(num == 9):
        return "SK"    
```

```python
pred = tf.argmax(layer_3, 1)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    while True:
        inp = input()
        
        if inp == '종료':
            break
        
        inp = inp.split(' ')
        inp[0] = KBO_num(inp[0])
        inp[1] = KBO_num(inp[1])
        
        if inp[0] < inp[1]:
            x = [[inp[0], inp[1]]]
        else:
            x = [[inp[1], inp[0]]]
        
        result = sess.run(pred, feed_dict={X:x})
        
        print("%s 우승!" % (num_to_team(x[0][result[0]])))
```
